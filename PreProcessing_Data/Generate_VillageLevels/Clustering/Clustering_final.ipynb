<font size=4 color="blue">Function to create the box plots. Here cluster_feat represents just the feature part of the feature list corresponding to that cluster (i.e. without the index part).</font>

```python
def create_box_plots(cluster_0, cluster_1, cluster_2):

    # TV, CYCLE, Television, 2W and 4W
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
    labels = ['RUD', 'INT', 'ADV']

    ########################################## BOX PLOT 0  ##########################################
    cluster_0_feat = [np.array([int(i[j]) for i,k in cluster_0]) for j in range(3)]
    
    input_0 = [cluster_0_feat[0],cluster_0_feat[1],cluster_0_feat[2]]

    # rectangular box plot
    bplot0 = axes[0].boxplot(input_0,
                             vert=True,  # vertical box alignment
                             patch_artist=True,  # fill with color
                             labels=labels)  # will be used to label x-ticks
    axes[0].set_title('CLUSTER 1')

    ########################################## BOX PLOT 1  ##########################################
    cluster_1_feat = [np.array([int(i[j]) for i,k in cluster_1]) for j in range(3)]
    
    input_1 = [cluster_1_feat[0],cluster_1_feat[1],cluster_1_feat[2]]

    bplot1 = axes[1].boxplot(input_1,
                             vert=True,  # vertical box alignment
                             patch_artist=True,  # fill with color
                             labels=labels)  # will be used to label x-ticks
    axes[1].set_title('CLUSTER 2')

    ########################################## BOX PLOT 2  ##########################################
    cluster_2_feat = [np.array([int(i[j]) for i,k in cluster_2]) for j in range(3)]
    
    input_2 = [cluster_2_feat[0],cluster_2_feat[1],cluster_2_feat[2]]

    bplot2 = axes[2].boxplot(input_2,
                             vert=True,  # vertical box alignment
                             patch_artist=True,  # fill with color
                             labels=labels)  # will be used to label x-ticks
    axes[2].set_title('CLUSTER 3')


    # fill with colors
    colors = ['pink', 'lightblue', 'lightgreen','red','blue']
    for bplot in (bplot0, bplot1, bplot2):
        for patch, color in zip(bplot['boxes'], colors):
            patch.set_facecolor(color)

    # adding horizontal grid lines
    for ax in axes:
        ax.yaxis.grid(True)
        ax.xaxis.grid(True)
        ax.set_xlabel('Main-Source-of-Water wise distribution')
        ax.set_ylabel('Observed % values')
        ax.set_ylim(0, 100)    

    plt.show()
    
    return cluster_0_feat, cluster_1_feat, cluster_2_feat
```

<font size=4 color="blue">Function to calculate the outliers using the Inter-Quartile range method, given a 1-D array of data (the variable "cluster"). The outliers are represented as indices in the cluster i.e. if an outlier has a value "i", then that means it is the i'th indexed value in the cluster list passed to the function.</font>

```python
def calc_outliers(cluster, outliers):

  # MEDIAN
  Q2 = np.median(cluster)

  # MEDIAN OF UPPER HALF
  Q3 = np.median(cluster[np.where(cluster > Q2)])

  # MEDIAN OF LOWER HALF
  if np.where(cluster < Q2)[0].shape[0] != 0:
    Q1 = np.median(cluster[np.where(cluster < Q2)])
  else:
    Q1 = 0.0

  # INTER-QUARTILE RANGE
  IQR = Q3-Q1

  # FILTER
  if np.where(cluster < Q2)[0].shape[0] != 0:
    outliers = outliers.union(set(np.where((cluster <= Q1 - 1.5*IQR) | (cluster >= Q3 + 1.5*IQR))[0].tolist()))
  else:
    outliers = outliers.union(set(np.where((cluster >= Q3 + 1.5*IQR))[0].tolist()))

  return outliers
```

<font size=4 color="blue">Function to calculate the outliers for each cluster by passing the feature list for each cluster to the function, calculating outliers for each type of asset(tv, tele, 2w and 4w) and combining the outliers for all assets to form the set of outliers for that cluster. outliers0 represents the outliers for cluster_0 and so on.</font>

```python
def all_outliers(outliers0, outliers1, outliers2, cluster_0_feat, cluster_1_feat, cluster_2_feat):

    # Calculate the set of outliers
    for i in range(3):
        outliers0 = calc_outliers(cluster_0_feat[i], outliers0)
        outliers1 = calc_outliers(cluster_1_feat[i], outliers1)
        outliers2 = calc_outliers(cluster_2_feat[i], outliers2)
    
    return outliers0, outliers1, outliers2
```

<font size=4 color="blue">Function to filter out the outliers from each cluster using the index values in the outlier lists</font>

```python
# Filter the clusters on the basis of outliers
def cluster_filter_on_outliers(cluster_0, cluster_1, cluster_2, outliers0, outliers1, outliers2):
    
    cluster_0_ind = []

    for i in range(len(cluster_0)):
      if i not in outliers0:
        cluster_0_ind += [i]

    cluster_0_new = cluster_0[cluster_0_ind].copy()

    cluster_1_ind = []

    for i in range(len(cluster_1)):
      if i not in outliers1:
        cluster_1_ind += [i]

    cluster_1_new = cluster_1[cluster_1_ind].copy()

    cluster_2_ind = []

    for i in range(len(cluster_2)):
      if i not in outliers2:
        cluster_2_ind += [i]

    cluster_2_new = cluster_2[cluster_2_ind].copy()

    return cluster_0_new, cluster_1_new, cluster_2_new
```

<font size=4 color="blue">Iteratively remove the outliers until threshold percentage(0.5%) of outliers reached. The variable names have the same meaning as above.</font>

```python
total_villages = len(feature_list)
outlier_percentage = 100

while outlier_percentage >= 0.5:
  # display the box plots
  cluster_0_feat, cluster_1_feat, cluster_2_feat = create_box_plots(cluster_0, cluster_1, cluster_2)

  outliers0 = set()
  outliers1 = set()
  outliers2 = set()

  # Calculate the outliers for this iteration 
  outliers0, outliers1, outliers2 = all_outliers(outliers0, outliers1, outliers2, cluster_0_feat, cluster_1_feat, cluster_2_feat)

  # Filter out the outliers from each cluster using the outliers calculated above
  cluster_0, cluster_1, cluster_2 = cluster_filter_on_outliers(cluster_0, cluster_1, cluster_2, outliers0, outliers1, outliers2)

  # Total number of outliers
  total_outliers = len(outliers0) + len(outliers1) + len(outliers2)

  print("Total number of outliers:",total_outliers)
  print("Total number of villages with outliers:",total_villages)
  print("Total number of villages without outliers:", np.union1d(np.union1d(cluster_0[:, 1], cluster_1[:, 1]), cluster_2[:, 1]).astype(int).shape[0])

  # Calculate the new percentage of the remaining outliers
  outlier_percentage = (total_outliers/total_villages) * 100
  total_villages -= total_outliers
```
