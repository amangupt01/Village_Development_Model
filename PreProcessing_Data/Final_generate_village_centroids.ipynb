# Generate Village Centroids

```python
import json
import os
from tqdm import tqdm
import pickle
import numpy as np  

geopath = '/content/drive/MyDrive/Village_level_new/state_json'
```

```python
def nested_len(lst):
  if len(lst) == 0:
    return 0

  if type(lst[0]) != type([]):
    return len(lst)

  return nested_len(lst[0])
```

```python
def dist(c1, c2):
  return ((c1[0]-c2[0])**2) + ((c1[1]-c2[1])**2)
```

```python
vill_geoinfo = {}

state_list = ['UP','MP','MH','RJ','BR','OR','JH','CG','GJ','AP','KA','PB','HR','TN','WB']


for state in tqdm(os.listdir(geopath)):
  # f = open(os.path.join(geopath,state), 'r')
  if state == 'pickle_files':
    continue
  state_name = state.split('.')[0]  
  if state_name not in state_list : 
    continue

  print(" state : "+state_name)  
  f = open(os.path.join(geopath,state), 'r')
  jsonstring = f.read()
  data = json.loads(jsonstring)

  for vill in data['features']:
    if vill['properties'].get('pc11_village_id'): 
      vill_id = np.int64(vill['properties']['pc11_village_id'])
    elif vill['properties'].get('pc11_villa'):
      vill_id = np.int64(vill['properties']['pc11_villa'])  
    else:
      continue  
    #if vill_id not in comb_dict:
    #  continue

    mean_coord = [0,0]

    if type(vill['geometry']['coordinates'][0][0][0]) == type([]):
      coord_list = vill['geometry']['coordinates'][0][0]
    else:
      coord_list = vill['geometry']['coordinates'][0]

    # for coord in coord_list:
    #   if coord[0] < 0.0 or coord[1] < 0.0:
    #     negative += 1
    #   mean_coord[0] += coord[0]
    #   mean_coord[1] += coord[1]

    coord_arr = np.array(coord_list)
    mean_coord = list(np.mean(coord_arr, axis=0))
    vill_geoinfo[vill_id] = mean_coord
```
