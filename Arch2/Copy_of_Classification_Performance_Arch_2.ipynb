```python
IN = pickle.load(open('/content/drive/MyDrive/Scripts_and_Data_Transfer/3.Arch_2_Procedure/5.Train_Regression_and_Classification_Models/Inputs/Household_Indicators_Input/household_input_2011_combined_new.pickle', 'rb'))
IN
```

```python
model_lis = ['BF', 'FC', 'MSW', 'ASSET', 'LIT']

def add1(a) :
  return a+1

arch1_df = IN.copy().drop(list(IN.columns), axis = 1)

for model_name in model_lis :
  test_list = IN.copy()[[model_name+'_0_OUT_1', model_name+'_0_OUT_2', model_name+'_0_OUT_3']]

  test_arr = test_list.copy().to_numpy()
  preds = list(map(np.argmax, test_arr))
  preds = list(map(add1, preds))
  arch1_df[model_name] = preds
```

```python
arch1_df
```
