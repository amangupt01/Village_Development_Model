```python
def random_forest_features(X_train, Y_train, param_grid):
  sel = RandomForestRegressor(n_estimators=param_grid['n_estimators'], 
                              max_features=param_grid['max_features'], 
                              min_samples_split=param_grid['min_samples_split'],
                              max_samples=param_grid['max_samples'], 
                              oob_score=param_grid['oob_score'], 
                              n_jobs=param_grid['n_jobs'], 
                              random_state=param_grid['random_state'])
  sel.fit(X_train, Y_train)
  return sel.feature_importances_
```
