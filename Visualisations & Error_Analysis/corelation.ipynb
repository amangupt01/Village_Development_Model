```python
import pandas as pd
```

```python
ntl_2001 = pd.read_pickle('ntl_data/data_2001.pickle')
ntl_2011 = pd.read_pickle('ntl_data/data_2011.pickle')
ntl_2019 = pd.read_pickle('ntl_data/data_2019.pickle')
```

```python
adi_2001 = pd.read_pickle('data/pred_2001.pkl')
adi_2011 = pd.read_pickle('data/pred_2011.pkl')
adi_2019 = pd.read_pickle('data/pred_2019.pkl')
```

```python
def change_index_type_to_int(df):
    return df.set_index(df.index.astype('int'))
```

```python
ntl_2001 = change_index_type_to_int(ntl_2001)
ntl_2011 = change_index_type_to_int(ntl_2011)
ntl_2019 = change_index_type_to_int(ntl_2019)
```

```python
adi_2001.head()
```

```python
ntl_2001.head()
```

```python
df_2001 = pd.merge(adi_2001, ntl_2001, left_index=True, right_index=True)
df_2011 = pd.merge(adi_2011, ntl_2011, left_index=True, right_index=True)
df_2019 = pd.merge(adi_2019, ntl_2019, left_index=True, right_index=True)
```

```python
data_01_11 = pd.merge(df_2001,df_2011,left_index=True,right_index=True,suffixes=("_2001","_2011"))
```

```python
data_01_11.head()
```

```python
data_01_11["Delta_ADI"] = (data_01_11["ADI_2011"] - data_01_11["ADI_2001"])
for col_name in ntl_2001.columns:
    data_01_11["Delta_"+ col_name] = (data_01_11[col_name+"_2011"] - data_01_11[col_name+"_2001"])
```

```python
final_01_11 = data_01_11[["Delta_ADI"]+["Delta_"+i for i in ntl_2001.columns]]
final_01_11.shape
```

```python
final_01_11 = final_01_11[final_01_11["Delta_ADI"]<0]
final_01_11.shape
```

```python
for col_name in ntl_2001.columns:
    print(col_name,final_01_11["Delta_ADI"].corr(final_01_11["Delta_"+col_name]),final_01_11[final_01_11["Delta_"+col_name]<0].shape[0])
# print(final_01_11["Delta_ADI"].corr(final_01_11["Delta_ADI"]))
```

```python
data_11_19 = pd.merge(df_2011,df_2019,left_index=True,right_index=True,suffixes=("_2011","_2019"))
data_11_19["Delta_ADI"] = (data_11_19["ADI_2019"] - data_11_19["ADI_2011"])
for col_name in ntl_2001.columns:
    data_11_19["Delta_"+ col_name] = (data_11_19[col_name+"_2019"] - data_11_19[col_name+"_2011"])
final_11_19 = data_11_19[["Delta_ADI"]+["Delta_"+i for i in ntl_2011.columns]]
final_11_19 = final_11_19[final_11_19["Delta_ADI"]<0]
for col_name in ntl_2011.columns:
    print(col_name,final_11_19["Delta_ADI"].corr(final_11_19["Delta_"+col_name]))
```
